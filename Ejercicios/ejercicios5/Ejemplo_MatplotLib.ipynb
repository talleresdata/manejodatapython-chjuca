```python
event_source = df[['event_source', 'session']].groupby(['event_source'], as_index=False).count()
event_source
```

```python
row= event_source.event_source.values
columns = event_source.session.values
```

```python
colors = ['#FF8000', '#6E6E6E']
```

```python
plt.bar(x = row, height= columns, color = colors)
plt.title('Cantidad de Eventos segun la fuente')
```

```python
lista = df.host.unique().tolist()
host = df.groupby(['host']).size().reset_index(name='count')
percentage = host['count'].tolist()
```
